```python
import numpy
import pandas
from sklearn import linear_model
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

data = load_iris()
dataset = pandas.DataFrame(data=data, columns=data.feature_names)
dataset['species'] = data.target
print(dataset["species"])
```

```python
# Map target integers to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
dataset['species'] = dataset['species'].map(species_map)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='viridis')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
